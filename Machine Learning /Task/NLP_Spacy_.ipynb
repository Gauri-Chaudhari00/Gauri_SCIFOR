```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("Machine learning (ML) is a branch of artificial intelligence (AI) and computer science that focuses on the using data and algorithms to enable AI to imitate the way that humans learn, gradually improving its accuracy.")
```

```python
#Tokenization
for token in doc:
  print(token.text)
```

```python
#POS tagging
for token in doc:
  print(token.text, token.pos_)
```

```python
#Dependency parsing
for token in doc:
    print(f"{token.text}: {token.dep_} -> {token.head.text}")
```

```python
#Visually display the dependencies
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Machine learning (ML) is a branch of artificial intelligence (AI) and computer science that focuses on the using data and algorithms to enable AI to imitate the way that humans learn, gradually improving its accuracy.")
sentence_spans = list(doc.sents)
displacy.serve(sentence_spans, style="dep")
```

```python
#Named Entity Recognition
doc = nlp("Machine learning is behind chatbots and predictive text, language translation apps, the shows Netflix suggests to you, and how your social media feeds are presented. ")
for ent in doc.ents:
  print(ent.text, ent.label_)
```

```python
#Lemmatization
for token in doc:
  print(token.text, token.lemma_)
```

```python
for sent in doc.sents:
  print(sent.text)
```
