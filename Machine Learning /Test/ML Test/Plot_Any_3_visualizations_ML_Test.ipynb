```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

# Create a small synthetic dataset with missing values
np.random.seed(20)
X = np.random.randn(20, 5)
y = np.random.randint(0, 2, 20)

# Introduce missing values in the dataset
X[np.random.choice([True, False], size=X.shape, p=[0.2, 0.8])] = np.nan

# Convert to DataFrame for convenience
X = pd.DataFrame(X, columns=[f'feature_{i}' for i in range(5)])
y = pd.Series(y, name='target')

print("Dataset with missing values:")
print(X)

# Impute missing values with mean
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

print("\nDataset after imputing missing values:")
print(X_imputed)

# Visualization 1: Correlation Heatmap
plt.figure(figsize=(8, 6))
corr_matrix = X_imputed.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Visualization 2: Histograms of Features
X_imputed.hist(bins=10, figsize=(10, 6))
plt.suptitle('Histograms of Features')
plt.show()

# Visualization 3: Count Plot of Target Variable
plt.figure(figsize=(6, 4))
sns.countplot(x=y)
plt.title('Count Plot of Target Variable')
plt.show()
```
