```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
#from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier,plot_tree

# Importing the data
df = pd.read_csv(r'/content/drive/MyDrive/Meta Scifor/Datasets/ipl2024 Matches.csv',)
df.head()
print(df)
```

```python
le = LabelEncoder()
df['id_n']= le.fit_transform (df['id'])
df['team1_n']= le.fit_transform (df['team1'])
df['team2_n']= le.fit_transform (df['team2'])
df['winner_n']= le.fit_transform (df['winner'])
df['player of match_n']= le.fit_transform (df['player of match'])

```

```python
df
```

```python
df= df.drop(['id'	,'team1', 'team2','winner',	'player of match'],axis='columns')
```

```python
df
```

```python
independent_variable = df.drop(['player of match_n'],axis='columns')
dependent_var= df['player of match_n']
```

```python
model= tree.DecisionTreeClassifier()
```

```python
model.fit(independent_variable, dependent_var)
```

```python
model.score(independent_variable, dependent_var)
```

```python
model.predict([[4,2,3,1]])
```

```python
plt.figure(figsize=(120,80))
plot_tree(model, filled= True, feature_names=df.columns,class_names=le.classes_)
plt.show()
```
